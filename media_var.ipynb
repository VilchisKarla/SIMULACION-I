```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate

# Parámetros del generador
a = 82
c = 69
m = 10**4
X0 = 5

def generador(a, c, m, X0):
    x = (a * X0 + c) % m
    return x

muestra = []
lista = []

for i in range(10000):
    nuevo = generador(a, c, m, X0)
    muestra.append([i, X0, nuevo / m])
    X0 = nuevo
    lista.append(nuevo / m)

encabezados = ["i", "x_i", "x_(i+1)"]
print(tabulate(muestra, headers=encabezados))

media = np.mean(lista)
varianza = np.var(lista)

print("Media:", media)
print("Varianza:", varianza)

# Función para calcular la CDF
def cdf(datos):
    x = np.sort(datos)  # Ordena los datos
    y = np.arange(1, len(datos) + 1) / len(datos)  # Genera la escala acumulativa
    return x, y

# Calcular la CDF
x, y = cdf(lista)

plt.figure(figsize=(12, 5))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(lista, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histograma de Xn/m")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")

# Gráfica de la CDF
plt.subplot(1, 2, 2)
plt.plot(x, y, marker='.', linestyle='none', color='b')
plt.title("Función de Distribución Acumulada (CDF)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia Acumulada")
plt.grid(True)

plt.show()
```
