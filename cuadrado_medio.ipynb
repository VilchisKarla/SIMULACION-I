# **Cuadrado Medio**

El método comienza tomando un número al azar "x0", de 2n cifras (originalmente los autores proponían 4 cifras) que al elevarlo al cuadradoresulta un número de hasta 4n cifras. Si es necesario se añaden ceros a la izquierdapara que el número resultante tenga exactamente 4n cifras. Sea x1 el número resultantede seleccionar las 2n cifras centrales de x02; el primer número aleatorio u1 se obtieneponiendo un punto decimal delante las 2n cifras de x1. A continuación x2 y u2 se generan a partir de x1 del mismo modo. Así sucesivamente.sucesivamente.


---



**Desventajas**


*   tiene una fuerte tendencia a degenerar a cero rápidamente
*   los números generados pueden repetirse cíclicamente después de una secuenciacorta.



```python
import math
import tabulate

def middle_square_method(seed, n):
  num_rand = []  # Guarda los números generados
  iterations = []  # guarda las iteraciones
  intervalo = None  # Inicializamos el intervalo para el ciclo
  for i in range(n):
    seed_str = str(seed) #convierte los digitos de la semilla en una cadena
    seed_str = seed_str.zfill(8) #rellena la cadena con ceros hasta que sean 8 digitos
    seed_squared = seed * seed  #eleva al cadrado la semilla
    seed_squared_str = str(seed_squared)  #convierte el cuadrado de la semilla en cadena
    seed_squared_str = seed_squared_str.zfill(8) #rellena con ceros la cadena parpara que sean 8 digitos
    middle_digits = seed_squared_str[2:6] ##toma los 4 digitos del centro

    # convierte los digitos medios en la siguiente semilla
    seed = int(middle_digits)
    num_rand.append(seed)  # guarda los numeros random en la lista

    iterations.append([i + 1, seed, seed_squared_str, middle_digits])

#detecta el ciclo
    if num_rand and (seed in num_rand[:-1] or 0 in num_rand):
      # si el ciclo es detectado, asigna la iteración correspondiente a 'intervalo'
      intervalo = i + 1
      break # cierra el ciclo

  return iterations, intervalo

encabezados = ["i", "Semilla", "Semilla²", "Dígitos Medios"]

seed = 1234
n = 100
# Imprimir el ciclo de vida si se encontró
if intervalo is not None:
    print(f"\033[1mEl ciclo de vida es:\033[0m {intervalo}")#imrpime en negritas el ciclo de vida
else:
    print("No se encontró un ciclo en las iteraciones dadas.")
# imprime la tabla con los valores generados
iterations_data, intervalo = middle_square_method(seed, n)
print(tabulate.tabulate(iterations_data, headers=encabezados, tablefmt="grid",
    numalign="center",
    stralign="center"))
```
