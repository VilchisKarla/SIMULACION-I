Los números generados corresponden siempre al residuo de la división de (aX_j +c) entre m, por lo que los valores a obtener siempre serán 0, 1, 2, ..., m-2, m-1. Los números obtenidos son determinísticos.


*   X0: Semilla  - - - - - - - - - - - - -  X0≥0

*   a: multiplicador - - - - - - - - - - a≥0

*   c: incremento - - - - - - - - - - - c≥0

*   m: módulo - - - - - - - - - - m>X0, m>a, m>c

secuencia de números aleatorios dada por:

**X_{j+1}=(aX_j +c)mod m**

En algunas sucesiones la semilla no siempre es el primer valor en repeetirse, esto ocurre cuando a y m no son primos relativos, esto implica que la semilla no vuelva a aparecer y se "degenere" la sucesión. Lo deseable sería tener sucesiones de números aleatorios no degeneradas.


---
Al dividir los números de la sucesión entre el módulo m obtenemos números aleatorios estandarizados.  
**R=X/m ∈ [0,1)**





```python
# TABLA DE GENERADOR CONGRUENCIAL
import tabulate #importa la paquetería para generar la tabla

# Parámetros del generador
a=81
c=89
m=100
X0=5 # Semilla
n = m+10 # Número de valores a generar

# Función del generador
def generador(a, c, m, X0):
    return (a * X0 + c) % m

# Inicialización de estructuras
muestra = [] #guarda en una lista los datos  que se vayan generando de i,X0,Xn/m
nuevo = []  # Lista para almacenar los valores generados de Xn
valor_visto = {} #diccionario de los numeros generados para vizualizar la repetición
intervalo = None  # Inicializamos intervalo

# Generación de números
for i in range(n):
    Xn = generador(a, c, m, X0)  # Genera el siguiente valor
    nuevo.append(Xn)  # Guarda en la lista los números generados
    muestra.append([i, X0, Xn / m])  # Guarda en la lista para mostrar los datos en la tabla

    # Comprobar si ya vimos este valor antes
    if X0 in valor_visto:
        valor_visto[X0].append(i)
        intervalo = i - valor_visto[X0][-2]
    else:
        valor_visto[X0] = [i]  # Registrar la primer ocurrencia

    X0 = Xn  # Actualizar para la siguiente iteración

# Imprimir la tabla con los valores generados
encabezados = ["i", "x_i", "x_{i+1}"]
print(tabulate.tabulate(muestra, headers=encabezados, tablefmt="fancy_grid",
    numalign="center",
    stralign="center"))

# Imprimir el ciclo de vida si se encontró
if intervalo is not None:
    print(f"\033[1mEl ciclo de vida es:\033[0m {intervalo}")
else:
    print("No se encontró un ciclo en las iteraciones dadas.")
```

```python
import matplotlib.pyplot as plt
from tabulate import tabulate

# Parámetros del generador
a=81
c=89
m=100
X0=5
#Función del generador
def generador(a, c, m, X0):
    x = (a * X0 + c) % m
    return x

muestra = []
lista = []
#ciclo para el generador
for i in range(100):
    nuevo = generador(a, c, m, X0)
    muestra.append([i, X0, f"{nuevo / m:.6f}"])
    X0 = nuevo
    lista.append(nuevo / m)
#Tabla de datos
encabezados = ["i", "x_i", "x_(i+1)"]
print(tabulate(muestra, headers=encabezados, tablefmt="grid",
    numalign="center",
    stralign="center",
    colalign=("center", "center", "center")))

plt.figure(figsize=(12, 5))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(lista, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histograma de Xn/m")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
#Secuencia generada
plt.subplot(1, 2, 2)
plt.plot(lista)
plt.title("Secuencia generada")
plt.xlabel("n")
plt.ylabel("Xn/m")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate

# Parámetros del generador
a = 82
c = 69
m = 10**4 #ayuda a determinar el ciclo de vida
X0 = 5 #punto inicial (SEMILLA)

def generador(a, c, m, X0):
    x = (a * X0 + c) % m
    return x

muestra = []#guardará nuestra tabla de elementos i,x_i y xn/m
lista = [] #guarda nuestro xn/m

for i in range(10000):
    nuevo = generador(a, c, m, X0)
    muestra.append([i, X0, nuevo / m])
    X0 = nuevo
    lista.append(nuevo / m)

encabezados = ["i", "x_i", "x_(i+1)"]
print(tabulate(muestra, headers=encabezados, tablefmt="grid",
    numalign="center",
    stralign="center",
    colalign=("center", "center", "center")))

media = np.mean(lista)
varianza = np.var(lista)

print("\033[1mMedia\033[0m", media)
print("\033[1mVarianza\033[0m", varianza)

# Función para calcular la CDF
def cdf(datos):
    x = np.sort(datos)  # Ordena los datos
    y = np.arange(1, len(datos) + 1) / len(datos)  # Genera la escala acumulativa
    return x, y

# Calcular la CDF
x, y = cdf(lista)

plt.figure(figsize=(12, 5))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(lista, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histograma de Xn/m")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")

# Gráfica de la CDF
plt.subplot(1, 2, 2)
plt.plot(x, y, marker='.', linestyle='none', color='b')
plt.title("Función de Distribución Acumulada (CDF)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia Acumulada")
plt.grid(True)

plt.show()
```
